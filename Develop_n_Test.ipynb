```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
import math
import time 
```

```python
# Parameters
R = 5  # Radius of the fixed circle
r = 2  # Radius of the rolling circle

xArray = []
yArray = []

xArrayII = []
yArrayII = []

xArrayIII = []
yArrayIII = []

for theta in np.arange(0, 15, 0.1):

    display.clear_output(wait=True)
    fig, ax = plt.subplots()
    plt.axis( [-20,20, -20,20])

    x = (R + r) * math.cos(theta) - r * math.cos((R + r) / r * theta)
    y = (R + r) * math.sin(theta) - r * math.sin((R + r) / r * theta)
    xArray.append(x)
    yArray.append(y)

# Small r Value effects
#       size(r) semicircle         size of twist     no. of rings
    xII = (R + (r)) * math.cos(theta+0.6) - (r) * math.cos((R + (r)) / (r) * (theta))
    yII = (R + (r)) * math.sin(theta+0.6) - (r) * math.sin((R + (r)) / (r) * (theta))
    xArrayII.append(xII)
    yArrayII.append(yII)

    xIII = (R + (r)) * math.cos(theta+1.2) - (r) * math.cos((R + (r)) / (r) * (theta))
    yIII = (R + (r)) * math.sin(theta+1.2) - (r) * math.sin((R + (r)) / (r) * (theta))
    xArrayIII.append(xIII)
    yArrayIII.append(yIII)


    plt.plot(xArray,yArray,'g')
    plt.plot(xArrayII,yArrayII,'g')
    plt.plot(xArrayIII,yArrayIII,'g')
    plt.show()
```

```python
import pygame
import numpy as np
from scipy.spatial import Delaunay
import time

# Initialize Pygame
pygame.init()

# Screen dimensions
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('Dynamic Delaunay Triangulation')

# Colors
black = (0, 0, 0)
white = (255, 255, 255)
blue = (0, 0, 255)
gray = (128, 128, 128)

# Generate random points
points = np.random.rand(30, 2)
points[:, 0] *= width
points[:, 1] *= height

# Function to update points
def update_points(points):
    points += np.random.randn(*points.shape)
    points[:, 0] = np.clip(points[:, 0], 0, width)
    points[:, 1] = np.clip(points[:, 1], 0, height)
    return points

# Main loop
running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Update points
    points = update_points(points)
    
    # Clear screen
    screen.fill(black)

    # Compute Delaunay triangulation
    tri = Delaunay(points)
    
    # Draw triangles
    for simplex in tri.simplices:
        pygame.draw.polygon(screen, gray, points[simplex], 1)
    
    # Draw points
    for point in points:
        pygame.draw.circle(screen, blue, (int(point[0]), int(point[1])), 3)
    
    # Update display
    pygame.display.flip()
    
    # Cap the frame rate
    clock.tick(30)

# Quit Pygame
pygame.quit()
```
