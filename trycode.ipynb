```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

width, height = 800, 800
max_iter = 256
x = np.linspace(-2, 1, width)
y = np.linspace(-1.5, 1.5, height)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y
Z = np.array([mandelbrot(c, max_iter) for c in C.flat]).reshape(C.shape)

plt.imshow(Z, extent=[-2, 1, -1.5, 1.5], cmap='hot')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Ellipse, Rectangle, Polygon, Arc, Wedge
from matplotlib.lines import Line2D
from matplotlib.text import Text

fig, ax = plt.subplots()

# Add various shapes
circle = Circle((0.3, 0.3), 0.1, color='blue')
ellipse = Ellipse((0.7, 0.7), width=0.2, height=0.1, angle=45, color='green')
rectangle = Rectangle((0.5, 0.5), width=0.2, height=0.1, color='red')
polygon = Polygon([[0.1, 0.1], [0.2, 0.3], [0.3, 0.1]], closed=True, color='purple')
arc = Arc((0.5, 0.2), width=0.4, height=0.2, angle=0, theta1=0, theta2=180, color='orange')
wedge = Wedge((0.2, 0.8), r=0.1, theta1=0, theta2=90, color='yellow')
line = Line2D([0, 1], [0, 1], color='black')
text = Text(x=0.5, y=0.9, text='Hello, World!', fontsize=12, color='magenta')

# Add shapes to the axes
ax.add_artist(circle)
ax.add_artist(ellipse)
ax.add_artist(rectangle)
ax.add_artist(polygon)
ax.add_artist(arc)
ax.add_artist(wedge)
ax.add_artist(line)
ax.add_artist(text)

ax.set_aspect('equal')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')  # Optional: turn off the axis
plt.show()
```
